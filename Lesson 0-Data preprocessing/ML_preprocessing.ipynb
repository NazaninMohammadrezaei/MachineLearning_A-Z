```python
'''
__ Author __
Nazanin Mohammadrezaii

__ Topic__
Data preprocessing

__ Date __
Sep 2021

'''

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

# Import the dataset
df = pd.read_csv('Data.csv')
X = df.iloc[:,:3].values
Y = df.iloc[:,3].values  
df.head()
```

```python
# Handle missing data
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan,strategy = 'mean')
imputer = imputer.fit(X[:,1:3])
X[:,1:3]= imputer .transform(X[:,1:3])

# Change categorical variable into numerical
from sklearn.preprocessing import LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

labelencoder_X = LabelEncoder()
X[:,0] = labelencoder_X.fit_transform(X[:,0])
onehotencoder = ColumnTransformer( [('one_hot_encoder', OneHotEncoder(categories='auto'), [0])], remainder='passthrough' )  # The column numbers to be transformed (here is [0] but can be [0, 1, 3])
X = onehotencoder.fit_transform(X)

labelencoder_Y = LabelEncoder()
Y = labelencoder_Y.fit_transform(Y)

# Split the data set into train set and test set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=0)

# Feature scaling
from sklearn.preprocessing import StandardScaler
s_sc = StandardScaler()
x_train = s_sc.fit_transform(x_train)
x_test = s_sc.fit_transform(x_test)
```
